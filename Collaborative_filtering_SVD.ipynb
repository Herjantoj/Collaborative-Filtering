## Filter dataset dengan Date Submitted antara rentang 26 Januari 2000 sampai dengan 31 Desember 2001

```python
jumlah_voting = len(df_reviews_filtered['AuthorId'])
jumlah_user = df_reviews_filtered['AuthorId'].nunique()
rata2_rating = jumlah_voting / jumlah_user

print("Data dalam rentang waktu 26 Januari 2000 - 31 Desember 2001")
print('Jumlah voting = {}'.format(jumlah_voting))
print('Jumlah user = {}'.format(jumlah_user))
print('rata-rata user melakukan rating = {} / {} = {}'.format(jumlah_voting,jumlah_user, rata2_rating))
```

## drop column

```python
# Drop attribut yang tidak akan digunakan
df_reviews_filtered.drop("AuthorName", axis=1, inplace=True)
df_reviews_filtered.drop("DateSubmitted", axis=1, inplace=True)
df_reviews_filtered.drop("DateModified", axis=1, inplace=True)
df_reviews_filtered.drop("ReviewId", axis=1, inplace=True)
df_reviews_filtered.drop("Review", axis=1, inplace=True)
```

# Algortima SVD

```python
!pip install scikit-surprise
```

```python
from surprise import accuracy, Dataset, SVD, Reader
from surprise.model_selection import KFold
```

```python
data = Dataset.load_from_df(df_reviews_filtered, Reader())
```

```python
# define a cross-validation iterator
kf = KFold(n_splits=5)

algo = SVD()

for i, (trainset, testset) in enumerate(kf.split(data)):

    # train and test algorithm.
    algo.fit(trainset)
    # predictions = algo.test(testset)

    # Compute and print Root Mean Squared Error
    print(f"RMSE for fold {i + 1}: {accuracy.rmse(algo.test(testset), verbose=True)}")
    print(f"MSE for fold {i + 1}: {accuracy.mse(algo.test(testset), verbose=True)}")
```

```python
# total recipe yang ada di dataset
all_recipe = pd.DataFrame(df_reviews_filtered['RecipeId'].unique(), columns=['RecipeId']) #jumlah film
all_recipe
```

```python
all_recipe.nunique()
```

```python
# recipe apa saja yang sudah dirating oleh AuthorId = 85528
rated = df_reviews_filtered[df_reviews_filtered.AuthorId == 85528]
rated.head()
```

```python
print('Recipe yang sudah dirating oleh user 85528 = {}'.format(rated['RecipeId'].nunique()))
```

```python
# Recipe yang belum di rating oleh AuthorId 85528
rated_list = pd.Series(rated['RecipeId'])
not_rated = all_recipe['RecipeId'].isin( rated_list)
not_rated = all_recipe[~not_rated]
not_rated.reset_index(inplace=True)
not_rated.drop(['index'], axis=1, inplace=True)
not_rated.head()
```

# Predict and Recomendation

```python
author_id = 85528
```

```python
# prediksi rating author dengan id 85528
score = pd.DataFrame([algo.predict(author_id, x).est for x in not_rated['RecipeId']], columns=['rating'])
score.head()
```

## Top-10 food recommendations (recipes) for Author with ID 85528

```python
result = not_rated.join(score['rating']).sort_values(by='rating', ascending=False)
result.head(10)
```
